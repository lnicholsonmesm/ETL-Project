```python
import imdb
```

```python
#call the imdb.IMDb function to get an access object through which IMDb data can be retrieved
ia = imdb.IMDb()
```

```python
#We can search through what subgenres are available in imdbpy
keywords = ia.search_keyword('romance')
keywords
```

```python
#We can get movies that are labeled as a specific subgenre
movies = ia.get_keyword('romance-episode')
movies
```

```python
#We can get the top 250 movies in imbdpy
top=ia.get_top250_movies()
top
```

```python
#Looped throught the top movies and made a list
ids = []

for i in range(250):
    ids.append(top[i].getID())
ids
```

```python
#We can get the rating for a specific movie
movie = ia.get_movie(ids[0])
movie.get('rating')
```

```python
#We can get the full cast for a movie
cast = movie.get('cast')
cast
```

```python
#We can get the plot for a movie
movie.get('plot')
```

```python
#Run through the cast of a movie and get the name of the first actor
actor = cast[0]
actor["name"]
```

```python
#Get that actor's ID
id_a = cast[0].getID()
id_a
```

```python
#Get that actor's current role
character = actor.currentRole
character
```

```python
#Get that actor's filmography
ia.get_person(id_a)["filmography"]
```
